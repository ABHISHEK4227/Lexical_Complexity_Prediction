# Import Libraries

```python
from dataset import Dataset
from features import Features_extracter as fe
from model import LCP
from scorer import scores
from scorer import saveOutput

from sklearn.model_selection import train_test_split
```

# For Multiple Word
Initialize the multi word dataset

```python
s = Dataset('multi')
f_train = fe()
x_train = f_train.preprocessing_m(s.df_train)
y_train = s.df_y_train

f_test = fe()
x_test = f_test.preprocessing_m(s.df_test)
y_test = s.df_y_test
```

# Train Validation *Split*

```python
x_train,x_val,y_train,y_val = train_test_split(x_train, y_train, test_size=0.2, random_state=42)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_val.shape)
print(y_val.shape)
```

# Train on Different Models

```python
m = LCP(x_train.shape[1])
m.trainall(x_train,y_train,x_val,y_val)
```

# Mean Stacking **[Final Model Used For Prediction]**
- Adaptive Boost
- Bagging Regressor 
- Gradient Boosting Regressor
- KNN

```python
y_pred1 = m.predict(x_test,7)
scores(y_test,y_pred1)
```

# Save the Results

```python
saveOutput(s.df_test_id,y_pred1,'Results_multi.txt')
```
